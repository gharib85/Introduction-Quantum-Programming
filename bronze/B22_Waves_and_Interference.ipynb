<h2> Waves and Interference </h2>

Here we focus on very simple form of waves to get some intutions on interference.

Sine waves (sinusoid) are nice tools for our purpose.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3*np.pi,3*np.pi,0.1)   # start,stop,step
y = np.sin(x)

yaxis = 0

plt.axes()

plt.plot(x,y,xaxis)
plt.show()
```
